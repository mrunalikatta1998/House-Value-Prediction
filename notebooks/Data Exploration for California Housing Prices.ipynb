```python
# from above datatypes we see 'ocean_proximity' column is non-numeric, hence we'll exclude that column and generate a correlation matrix
numeric_df = dataFrame.drop(columns=['ocean_proximity'])

# Correlation matrix
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(12, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='0.2f')
plt.title('Correlation Matrix')
plt.show()
```

- From the above correlation matrix we can see the relationship between all the features of the housing dataset.
- The Correlation coefficients range from -1 to 1.
- 1 indicates a perfect positive linear relationship.
- -1 indicates a perfect negative linear relationship.
- 0 indicates no linear relationship.
- Cool Colors like blue indicate a negative correlation.
- Warm Colors like red indicate a positive correlation.
- The intensity of the color represents the strength of the correlation. Darker shades usually denote stronger correlations.
- Median income shows a strong positive correlation with median house value. This indicates that areas with higher incomes tend to have more expensive houses.
- We see negative correlation between features like latitude and longitude.

```python
# Scatter plot for Total rooms vs Population
plt.figure(figsize=(10, 6))
plt.scatter(dataFrame['total_rooms'], dataFrame['population'], alpha=0.5)
plt.title('Total Rooms vs. Population')
plt.xlabel('Total Rooms')
plt.ylabel('Population')
plt.grid(True,alpha=0.4,linestyle='--')
plt.show()
```

```python
# bar plot for average housing median age by ocean proximity
avg_housing_age = dataFrame.groupby('ocean_proximity')['housing_median_age'].mean().sort_values()

plt.figure(figsize=(10, 6))
avg_housing_age.plot(kind='bar', color='mediumseagreen')
plt.title('Average Housing Age by Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Average Housing Age (Years)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.4, linestyle='--')
plt.show()
```

- The above bar plot tells the average housing age for the ocean proximity categories.
- The "ISLAND" category has the highest average housing age out of all the 5 categories, followed by "NEAR BAY" category indicating that people above age 35 prefer living here.
- The categories <1H OCEAN, NEAR OCEAN, and INLAND categories typically exhibit lower average housing ages when compared to ISLAND and NEAR BAY.
- The high housing prices near the ocean might be linked to the challenge younger populations face in affording such homes.

```python
from sklearn.linear_model import LinearRegression
from math import sqrt
from sklearn.metrics import mean_squared_error, r2_score

# Train Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict and evaluate
lr_preds = lr_model.predict(X_test)
mse = mean_squared_error(y_test, lr_preds)
rmse = sqrt(mse)
r2 = r2_score(y_test, lr_preds)

print("ðŸ“‰ Linear Regression Results:")
print(f"RMSE: {rmse:.2f}")
print(f"RÂ² Score: {r2:.4f}")
```
